Código de Aprendizado Supervisionado

```python
#Importando as bibliotecas

from sklearn import tree
import pandas as pd #Biblioteca para manipulação do excel
from sklearn.tree import DecisionTreeClassifier
```

```python
#Atribuindo o nosso excel base a uma variavel
df = pd.read_excel(r"\Users\Bruno\Documents\Github\Estudos de ML\qualidade_filmes.xlsx")

#Mostra os dados da planilha
df
```

```python
#Atribuindo a caracteristicas do resultado a uma variavel
caracteristicas = ["enredo", "plot", "elenco"]

#Atribuindo todas as caracteristica a uma variavel
x = df[caracteristicas]

#Atribuindo o resultado para uma variavel
resultado = df["qualidade filme"]


modelo = tree.DecisionTreeClassifier()
modelo.fit(x, resultado)

modelo.predict([[1,0,0]])
```

Código de Aprendizado Não Supervisionado

```python
from sklearn.cluster import KMeans
import pandas as pd

dados = {
    "idade": [18,22,38,23,33],
    "salario": [1000,200,5000,600,3000]
}

df = pd.DataFrame(dados)

x = df[["idade", "salario"]]

modelo = KMeans(n_clusters=4, random_state=42)

modelo.fit(x)

df["grupo"] = modelo.labels_
print(df)
```
